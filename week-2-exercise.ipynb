```python
while True:
    first_number = input("\nFirst number: ")
    if first_number == 'q' :
        break
    second_number = input("Second number: ")
    if second_number == 'q' :
        break
    try:
        answer = int(first_number) / int(second_number)
    except ZeroDivisionError:
        print("You can devide by 0!")
    else:
        print(answer)
```

```python
# write a function that takes from the user the file name, the user might sent a wrong file name, if so it should not break, it should print the content of the file out.
```

```python
def read_file():
    file_name = input("Enter the file name: ")
    try:
        with open(file_name, "r") as f:
            content = f.read()
            print("File content:\n")
            print(content)
    except FileNotFoundError:
        print(f"Error: The file '{file_name}' does not exist.")
    except Exception as e:
        print(f"An error occurred: {e}")

read_file()
```
