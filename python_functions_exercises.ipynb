```python
# 1. Print "Hello, World!"
def hello_world():
    """Prints Hello, World!"""
    print("Hello, World!")
```

```python
# 2. Greet a user by name
def greet(name):
    """Prints a greeting for the given name"""
    print(f"Hello, {name}!")
```

```python
# 3. Square of a number
def square(number):
    """Returns the square of a number"""
    return number ** 2
```

```python
# 4. Sum of two numbers
def add(a, b):
    """Returns the sum of two numbers"""
    return a + b
```

```python
# 5. Maximum of two numbers
def maximum(a, b):
    """Returns the maximum of two numbers"""
    return max(a, b)
```

```python
# 6. Check even or odd
def even_or_odd(number):
    """Checks if a number is even or odd"""
    return "Even" if number % 2 == 0 else "Odd"
```

```python
# 7. Length of a string
def string_length(text):
    """Returns the length of a string"""
    return len(text)
```

```python
# 8. Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    """Converts Celsius to Fahrenheit"""
    return (celsius * 9/5) + 32
```

```python
# 9. Average of a list
def average(numbers):
    """Returns the average of a list of numbers"""
    return sum(numbers) / len(numbers) if numbers else 0
```

```python
# 10. Count vowels in a string
def count_vowels(text):
    """Counts vowels in a string"""
    vowels = "aeiouAEIOU"
    return sum(1 for char in text if char in vowels)
```

```python
# 11. Unique values in a list
def unique_list(items):
    """Returns a list of unique values"""
    return list(set(items))
```

```python
# 12. Palindrome check
def is_palindrome(word):
    """Checks if a word is a palindrome"""
    return word == word[::-1]
```

```python
# 13. Simple interest
def simple_interest(principal, rate, time):
    """Calculates simple interest"""
    return (principal * rate * time) / 100
```

```python
# 14. Function with default tax rate
def calculate_tax(amount, tax_rate=0.075):
    """Calculates tax with a default tax rate"""
    return amount * tax_rate
```

```python
# 15. Factorial of a number
def factorial(n):
    """Returns the factorial of a number"""
    if n < 0:
        return None
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result
```

```python
# 16. Sum of any number of arguments
def sum_all(*args):
    """Returns the sum of all arguments"""
    return sum(args)
```

```python
# 17. Minimum value in a list
def minimum(numbers):
    """Returns the minimum value in a list"""
    return min(numbers)
```

```python
# 18. Convert minutes to hours and minutes
def minutes_to_hours(minutes):
    """Converts minutes to hours and minutes"""
    hours = minutes // 60
    remaining_minutes = minutes % 60
    return hours, remaining_minutes
```

```python
# 19. Welcome message
def welcome(username):
    """Prints a welcome message"""
    print(f"Welcome, {username}!")
```

```python
# 20. Area of a rectangle
def rectangle_area(length, width):
    """Returns the area of a rectangle"""
    return length * width
```

```python
# 21. Largest number in a list
def largest_number(numbers):
    """Returns the largest number in a list"""
    return max(numbers)
```

```python
# 22. Positive, negative, or zero
def number_sign(number):
    """Checks if a number is positive, negative, or zero"""
    if number > 0:
        return "Positive"
    elif number < 0:
        return "Negative"
    return "Zero"
```

```python
# 23. Reverse a string
def reverse_string(text):
    """Returns the reverse of a string"""
    return text[::-1]
```

```python
# 24. Power of a number
def power(base, exponent):
    """Calculates the power of a number"""
    return base ** exponent
```

```python
# 25. Grade calculation
def calculate_grade(score):
    """Returns grade based on default grading scale"""
    if score >= 70:
        return "A"
    elif score >= 60:
        return "B"
    elif score >= 50:
        return "C"
    elif score >= 45:
        return "D"
    return "F"
```

```python
# 26. Count words in a sentence
def word_count(sentence):
    """Counts words in a sentence"""
    return len(sentence.split())
```

```python
# 27. Prime number check
def is_prime(number):
    """Checks if a number is prime"""
    if number <= 1:
        return False
    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            return False
    return True
```

```python
# 28. Sum of digits
def sum_of_digits(number):
    """Returns the sum of digits of a number"""
    return sum(int(digit) for digit in str(abs(number)))
```

```python
# 29. Apply default discount
def apply_discount(prices, discount=0.10):
    """Applies a default discount to prices"""
    return [price * (1 - discount) for price in prices]
```

```python
# 30. Format full name
def full_name(first_name, last_name):
    """Returns formatted full name"""
    return f"{first_name.strip().title()} {last_name.strip().title()}"
```
